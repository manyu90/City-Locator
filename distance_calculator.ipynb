```python
from haversine import haversine
import pandas as pd
import numpy as np
```

```python
path_to_excel_city_file = './worldcities.xlsx'
city_df = pd.read_excel(path_to_excel_city_file,sheet_name='Sheet1')
```

```python
city_coords_dict = {}
for index,row in city_df.iterrows():
    print (index,row['city_ascii'],row['lat'],row['lng'])
    city_coords_dict[row['city_ascii']]=(row['lat'],row['lng'])
    
```

```python
city_coords_dict['London']
```

```python
class City():
    def __init__(self,name,lat,lng):
        self.name = name
        self.lat = lat
        self.lng = lng
        self.neighbours = {}
    def add_neighbour(self,neighbour,unit='mi'):
        coords_this_city = (self.lat,self.lng)
        coords_neighbour = (neighbour.lat,neighbour.lng)
        distance = haversine(coords_this_city,coords_neighbour,unit=unit)
        self.neighbours[neighbour.name] = distance
        
        
```

```python
def get_distance(city1,city2):
    assert(city1 in city_coords_dict)
    assert(city2 in city_coords_dict)
    lat1,long1 = city_coords_dict[city1]
    lat2,long2 = city_coords_dict[city2]
    return haversine((lat1,long1),(lat2,long2),unit='mi')

    
```

```python
distances = np.zeros((len(city_coords_dict),len(city_coords_dict)))
cities_list = city_coords_dict.keys()
        
```

```python
for i in range(len(cities_list)):
    for j in range(i,len(cities_list)):
        distance = get_distance(cities_list[i],cities_list[j])
        distances[i,j] = distance
        distances[j,i] = distance
    print("City {} is done: \n".format(str(i)))
    
    
```

```python
distances_df = pd.DataFrame(distances,index=cities_list,columns=cities_list)
```

```python
#distances_df = pd.read_pickle('./distances_dataframe.pkl)
def get_best_match_cities(city,distance_range,tolerance=10,cities_dataframe=None):
    assert(city in distances_df)
    possible_cities = []
    for other_city in distances_df[city].keys():
        dist = distances_df[city][other_city]
        if dist>distance_range-tolerance and dist<distance_range+tolerance :
            possible_cities.append((other_city,dist))
    if cities_dataframe is not None:
        cities_with_pop = []
        for elem in possible_cities:
            possible_city,possible_dist = elem[0],elem[1]
            population = max(cities_dataframe.loc[cities_dataframe['city_ascii']==possible_city]['population'].values)
            if np.isnan(population):
                population=0
            cities_with_pop.append((possible_city,possible_dist,population))
        cities_with_pop = sorted(cities_with_pop,key=lambda x: -1*x[2])
        possible_cities = cities_with_pop
    return(possible_cities)
            
```

```python
get_best_match_cities('San Francisco',6650,tolerance=50,cities_dataframe=city_df)
```
